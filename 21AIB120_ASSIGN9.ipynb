```python
from PIL import Image
import numpy as np

# Define a function to generate a Gaussian filter
def generate_gaussian_filter(sigma, filter_shape):
    m, n = filter_shape
    m_half, n_half = m // 2, n // 2
    gaussian_filter = np.zeros((m, n), np.float32)
    for y in range(-m_half, m_half):
        for x in range(-n_half, n_half):
            normal = 1 / (2.0 * np.pi * sigma**2.0)
            exp_term = np.exp(-(x**2.0 + y**2.0) / (2.0 * sigma**2.0))
            gaussian_filter[y + m_half, x + n_half] = normal * exp_term
    return gaussian_filter

# Define a function to perform convolution
def convolution(image, kernel):
    m_i, n_i, c_i = image.shape
    m_k, n_k = kernel.shape
    y_strides = m_i - m_k + 1
    x_strides = n_i - n_k + 1
    img = image.copy()
    output_shape = (m_i - m_k + 1, n_i - n_k + 1, c_i)
    output = np.zeros(output_shape, dtype=np.float32)
    count = 0
    output_tmp = output.reshape((output_shape[0] * output_shape[1], output_shape[2]))
    for i in range(y_strides):
        for j in range(x_strides):
            for c in range(c_i):
                sub_matrix = img[i:i + m_k, j:j + n_k, c]
                output_tmp[count, c] = np.sum(sub_matrix * kernel)
            count += 1
    output = output_tmp.reshape(output_shape)
    return output

# Define the standard deviation and filter shape for the Gaussian filter
sigma = 1.0
filter_shape = (5, 5)  # Adjust the filter shape as needed

# Directory where your images are located
image_dir = 'ResizedDataset/Train/Train/Healthy/'  # Change this to your dataset directory

# Output directory for the blurred images
output_dir_1 = 'Blurred_Dataset/Train/Train/Healthy'  # Create this directory if it doesn't exist

# Create the output directory if it doesn't exist
os.makedirs(output_dir_1, exist_ok=True)

# Get a list of image file names in the directory
image_files = os.listdir(image_dir)

# Process each image in the dataset
for image_file in image_files:
    # Load an image
    image_path = os.path.join(image_dir, image_file)
    image = Image.open(image_path)

    # Generate the Gaussian filter kernel
    gaussian_filter = generate_gaussian_filter(sigma, filter_shape)

    # Convert the image to a NumPy array
    image_array = np.array(image)

    # Perform convolution to apply Gaussian blur
    blurred_image_array = convolution(image_array, gaussian_filter)

    # Create a PIL image from the NumPy array
    blurred_image = Image.fromarray(blurred_image_array.astype(np.uint8))

    # Save the resulting image
    output_path = os.path.join(output_dir_1, image_file)
    blurred_image.save(output_path)

    print(f"Blurred and saved: {image_file}")
```

```python
import cv2
import numpy as np
import os

# Directory containing the HSV images
input_directory = 'HSV_Image_Dataset\Train\Train\Healthy'

# Create a directory to store the segmented images
output_directory = 'Watershed_Segmented_Images/Train/Train/Healthy'
os.makedirs(output_directory, exist_ok=True)  # Create the directory if it doesn't exist

# List the files in the input directory
image_files = os.listdir(input_directory)

# Function to apply Watershed segmentation
def watershed_segmentation(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to create a binary mask
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Morphological operations to clean up the mask
    kernel = np.ones((3, 3), np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)
    sure_bg = cv2.dilate(opening, kernel, iterations=3)

    # Distance transform
    dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
    _, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

    # Finding unknown region
    sure_fg = np.uint8(sure_fg)#sure_fg is a binary image with white regions representing the foreground.
    unknown = cv2.subtract(sure_bg, sure_fg)

    # Label markers
    _, markers = cv2.connectedComponents(sure_fg)
    markers = markers + 1
    markers[unknown == 255] = 0

    # Apply Watershed segmentation
    cv2.watershed(image, markers)
    image[markers == -1] = [0, 0, 255]

    return image

# Process and save segmented versions of the images using Watershed segmentation
for image_file in image_files:
    image_path = os.path.join(input_directory, image_file)
    hsv_image = cv2.imread(image_path)

    # Apply Watershed segmentation
    segmented_image = watershed_segmentation(hsv_image)

    # Save the segmented image in the separate directory
    output_path = os.path.join(output_directory, f'{image_file}_segmented_watershed.jpg')
    cv2.imwrite(output_path, segmented_image)

    print(f"Saved Watershed segmented version of {image_file} to {output_path}")
```
